```julia
function Calibration_up(start, stop, wedge)
    global beta, psi, alpha_e, alpha_s, delta_e, delta_s
    theta_e = wedge[1]
    theta_s = wedge[2]

    # Load data for calibration

    gr_tfp = [0; TFP[2:end] ./ TFP[1:end-1] .- 1]
    gr_en = [0; EN[2:end] ./ EN[1:end-1] .- 1]
    gr_ps = [0; p_Scpi[2:end] ./ p_Scpi[1:end-1] .- 1]
    gr_pe = [0; p_Ecpi[2:end] ./ p_Ecpi[1:end-1] .- 1]

    # Choose calibration years
    t1 = 1 + start - 1952
    t2 = 1 + stop - 1952

    cy = 1 - mean(I_E[t1:t2] ./ Y[t1:t2] .+ I_S[t1:t2] ./ Y[t1:t2])
    ksy = mean(K_S[t1:t2] ./ Y[t1:t2])
    key = mean(K_E[t1:t2] ./ Y[t1:t2])
    h = mean(EL[t1:t2] ./ EN[t1:t2])
    ga = log(TFP[t2] / TFP[t1]) / (t2 - t1)
    gn = log(EN[t2] / EN[t1]) / (t2 - t1)
    gps = log(p_Scpi[t2] / p_Scpi[t1]) / (t2 - t1)
    gpe = log(p_Ecpi[t2] / p_Ecpi[t1]) / (t2 - t1)

    # Assume either
    beta_assume = 1 / 1.07  # or
    alpha_tot = 0.4  # <--- This for now

    # x = alpha_e, alpha_s, beta, psi
    fn_sol = x -> cal_fn_up(x, delta_e, delta_s, cy, ksy, key, h, ga, gn, gps, gpe, alpha_tot, beta_assume, theta_e, theta_s)
    i=0
    i=1+i
    println("i = ", i)
    x0 = [0.2, 0.2, 0.8, 0.5, 0.8]
    result = Optim.optimize(fn_sol, x0, Optim.BFGS())

    sol = Optim.minimizer(result)

    alpha_e = sol[1]
    alpha_s = sol[2]
    beta = sol[3]
    psi = sol[4]

    g = (ga - alpha_e * gpe - gps * alpha_s) / (1 - alpha_e - alpha_s)

    return [sol..., g, ga, gn, gps, gpe]
end
```

```julia
function cal_fn_up(x, delta_e, delta_s, cy, ksy, key, h, ga, gn, gps, gpe, alpha_tot, beta_assume, theta_e, theta_s)
    g = (ga - x[1] * gpe - gps * x[2]) / (1 - x[1] - x[2])

    F = zeros(5)
    F[1] = x[4] * h / (1 - h) * x[5] - (1 - x[1] - x[2])
    F[2] = x[2] - ksy * exp(-gn) * (exp(g) * theta_s / x[3] - (1 - delta_s) * theta_s * exp(gps))
    F[3] = x[1] - key * exp(-gn) * (exp(g) * theta_e / x[3] - (1 - delta_e) * theta_e * exp(gpe))
    F[4] = x[2] + x[1] - alpha_tot
    #println("F[4] = ", F[4])
    
    # Alternative restriction
    # F[4] = x[3] - beta_assume

    # F(5) = 1 - exp(gn)*x(5) - key*theta_e*(exp(g+gn)-exp(gpe)*(1-delta_e)) -
    #    ksy*theta_s*(exp(g+gn)-exp(gps)*(1-delta_s));

    F[5] = 1 - x[5] - key * theta_e * (exp(g) - exp(gpe - gn) * (1 - delta_e)) -
        ksy * theta_s * (exp(g) - exp(gps - gn) * (1 - delta_s))

    return sum(abs2, F)
end
```
